```python


# Remove rows containing NaN or infinity values from the data
X = X[~np.isnan(X).any(axis=1)]
y = y[~np.isnan(y).any(axis=1)]
X_test = X_test[~np.isnan(X_test).any(axis=1)]
y_test = y_test[~np.isnan(y_test).any(axis=1)]
X = X[~np.isinf(X).any(axis=1)]
y = y[~np.isinf(y).any(axis=1)]
X_test = X_test[~np.isinf(X_test).any(axis=1)]
y_test = y_test[~np.isinf(y_test).any(axis=1)]

# Train the model and make predictions
clf = RandomForestClassifier()
clf.fit(X, y)
y_pred = clf.predict(X_test)

# Calculate the accuracy and precision of the model
accuracy = accuracy_score(y_test,y_pred)
# precision = precision_score(y_test,y_pred, average='weighted') 0.9512 0.9268
# precision = precision_score(y_test, y_pred, average='macro') 0.9512 0.9268
precision = precision_score(y_test, y_pred, average='micro') # 0.9512 0.9512
print("Accuracy: ", accuracy)
print("Precision: ", precision)
```
